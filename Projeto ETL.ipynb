1. EXTRACT 


```python
import google.generativeai as genai
import json

# Cole sua API Key do Google aqui 
GEMINI_API_KEY = "SUA_API_AQUI"
genai.configure(api_key=GEMINI_API_KEY)

# Lista de usuários (
users = [
    {"id": 1, "name": "Carlos Silva",   "news": []},
    {"id": 2, "name": "Ana Oliveira",   "news": []},
    {"id": 3, "name": "Beatriz Santos", "news": []},
    {"id": 4, "name": "Daniel Souza",   "news": []},
    {"id": 5, "name": "Fernanda Lima",  "news": []}
]
```

2. TRANSFORM (Gerando mensagens com Gemini)


```python
def generate_ai_news(user):
  # Configurando o modelo
  model = genai.GenerativeModel("gemini-2.5-flash")

  prompt = (
      f"Você é um especialista em marketing bancário. "
      f"Crie uma mensagem curta e impactante para {user['name']} "
      f"sobre a importância dos investimentos (máximo de 100 caracteres)."
      f"Me dê apenas uma sugestão de mensagem, não precisa me dar várias opções"
  )

  try:
    response = model.generate_content(prompt)
    return response.text.strip()
  except Exception as e:
    print(f"Erro ao gerar para {user['name']}: {e}")
    return "Invista hoje para garantir seu futuro!"

# Executando a transformação para cada usuário
print("Gerando mensagens com IA...")
for user in users:
  news = generate_ai_news(user)
  print(f" > {user['name']}: {news}")

  user['news'].append({
      "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
      "description": news
  })
```

3. LOAD (Salvando o resultado)

```python
with open("Cleint_Result.json", "w", encoding="utf-8") as f:
    json.dump(users, f, ensure_ascii=False, indent=4)

print("\nETL Finalizado! O arquivo 'Cleint_Result.json' foi gerado com sucesso.")
```
